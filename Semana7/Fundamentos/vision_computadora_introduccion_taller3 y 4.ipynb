```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
# Load an image using OpenCV
image = cv2.imread('/content/Lobo.JPG')
# Display the image
cv2_imshow(image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#obtener rotacion imagen
height,widh = image.shape[:2]

center = (widh/2,height/2)
center
```

```python
#definir la matriz de translacion
tx, ty =100, 50
M = np.float32([[1,0,tx],[0,1,ty]])
#aplicar la translacion
translated = cv2.warpAffine(image,M,(image.shape[1],image.shape[0]))
cv2_imshow(translated)
   # Esperar a que el usuario presione una tecla
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#deteccion de bordes

image= cv2.imread('Lobo.JPG',cv2.IMREAD_GRAYSCALE)

#aplicar operador sobel

sobelx= cv2.Sobel(image,cv2.CV_64F,1,0,ksize=5)
sobely= cv2.Sobel(image,cv2.CV_64F,0,1,ksize=5)
#combinar
edges=cv2.magnitude(sobelx,sobely)
edges = cv2.normalize(edges ,None,0,255,cv2.NORM_MINMAX,dtype=cv2.CV_8U)
cv2_imshow(edges)

```

```python
# Load an image using OpenCV
image = cv2.imread('/content/Gato.JPG')
# Display the image
#cv2_imshow(image)
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
#deteccion derostros
faceClassif=cv2.CascadeClassifier('haarcascade_frontalcatface.xml')
faces = faceClassif.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=5,minSize=(30,30),maxSize=(200,200))
#faceClassif.detectMultiScale(gris, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
print (faces)
#dibujar rectangulos
for (x,y,w,h) in faces:
  cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)
  cv2_imshow(image)

```
