```python
image_path ='C:/Users/William Wallace/Desktop/Lobo.jpg'
image =skimage.io.imread(image_path)
skimage.io.imshow(image)
skimage.io.show()
```

```python
rotated_image = skimage.transform.rotate(image, angle=90)
skimage.io.imshow(rotated_image)
skimage.io.show()
```

```python
gaussian = skimage.filters.gaussian(image, sigma=3)
skimage.io.imshow(gaussian)
skimage.io.show()
```

```python
# Realizar la reflexin horizontal
flipped_image = np.fliplr(image)

# Mostrar la imagen original y la reflejada
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image)
axes[0].set_title("Imagen Original")
axes[0].axis('off')

axes[1].imshow(flipped_image)
axes[1].set_title("Imagen Reflejada Horizontalmente")
axes[1].axis('off')

plt.show()
```

```python
# Aplicar el filtro de Sobel
sobel_edges = skimage.filters.sobel(image)
image_gray = skimage.color.rgb2gray(image)
# Mostrar la imagen original y el resultado del filtro de Sobel
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(image_gray, cmap='gray')
axes[0].set_title("Imagen en Escala de Grises")
axes[0].axis('off')

axes[1].imshow(sobel_edges, cmap='gray')
axes[1].set_title("Bordes con Filtro de Sobel")
axes[1].axis('off')

plt.show()
```
